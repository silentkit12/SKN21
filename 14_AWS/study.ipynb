구문오류와 예외
- 구문 오류 : 괄호의 개수, 들여쓰기의 문제등으로 프로그램 실행 전에 발생하는 오류
  - 해결하지 못하면 아예 실행히 되지 않음
            - SyntaxError
- 예외 : 런타임 오류, 실행중에 발생하는 오류 

예외처리
- 예외를 해결하는 것을 모두 예외처리라고 부름.
- 조건문/ try 구문을 사용하는 방법이 있음

```python
# 조건문을 통한 예외처리 : 기본 예외 처리

number_input_a = int(input("정수 입력>"))

print("원의 반지름", number_input_a)
print("원의 둘레:", 2*3.14*number_input_a)
print("원의 넒이:", 3.14*number_input_a*number_input_a)

# input에 정수가 들어가지 않는다면 예외가 발생함
```

```python
#위와 같은 상황을 방지하기 위해 "예외처리"를 함.

user_input_a = input("정수 입력>")

if user_input_a.isdigit():
    number_input_a = int(user_input_a)

    print("원의 반지름", number_input_a)
    print("원의 둘레:", 2*3.14*number_input_a)
    print("원의 넒이:", 3.14*number_input_a*number_input_a)
else:
    print("정수를 입력하지 않았습니다.")
```

```python
#try except 구문 : 예외처리를 위한 구문
#   try : 예외가 발생할 가능성이 있는 코드
#   except : 예외가 발생했을 때 실행할 코드

try:
    user_input_a = input("정수 입력>")

    print("원의 반지름", number_input_a)
    print("원의 둘레:", 2*3.14*number_input_a)
    print("원의 넒이:", 3.14*number_input_a*number_input_a)

except:
    print("무언가 잘못 되었습니다.") # 예외가 발생했을 때 실행할 구문.
```
